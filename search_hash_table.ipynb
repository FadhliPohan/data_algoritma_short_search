```python
import json
import time

# FILE_PATH = "data_pembelian_10000.json"
FILE_PATH = "data_dummy_pelanggan.json"

with open(FILE_PATH, 'r', encoding='utf-8') as file:
    data = json.load(file)

    len(data)
```

```python
from operator import itemgetter

# Membuat hash table untuk setiap jenis pencarian
def create_hash_table(data, key_field):
    hash_table = {}
    for record in data:
        # Jika key adalah angka, ubah menjadi string
        key = str(record[key_field]).strip().lower() if isinstance(record[key_field], str) else str(record[key_field])
        if key not in hash_table:
            hash_table[key] = []
        hash_table[key].append(record)
    return hash_table
```

```python
# Membuat hash table untuk pencarian berdasarkan bulan-tahun
def create_hash_table_by_bulanan(data):
    hash_table = {}
    for record in data:
        # Ekstrak bulan dan tahun dari tanggal
        bulan_tahun = record["tanggal_pembelian"][-7:]  # format: dd-mm-yyyy, mengambil mm-yyyy
        if bulan_tahun not in hash_table:
            hash_table[bulan_tahun] = []
        hash_table[bulan_tahun].append(record)
    return hash_table

# Membuat hash table untuk pencarian bulan-tahun
hash_table_by_bulanan = create_hash_table_by_bulanan(data)
```

```python
# Membuat hash table untuk setiap jenis data pencarian
hash_table_by_id = create_hash_table(data, "id")
hash_table_by_nama = create_hash_table(data, "nama_pelanggan")
hash_table_by_tanggal = create_hash_table(data, "tanggal_pembelian")
hash_table_by_jumlah = create_hash_table(data, "jumlah_pembelian")
hash_table_by_bulanan = create_hash_table_by_bulanan(data)
```

```python
def cari_data():
    print("Jenis pencarian yang tersedia:")
    print("1. id")
    print("2. nama")
    print("3. tanggal")
    print("4. jumlah")
    print("5. bulanan")

    jenis = input("Masukkan jenis pencarian (id/nama/tanggal/jumlah/bulanan): ").strip().lower()

    if jenis == "id":
        target = input("Masukkan ID (misal: t0001): ").strip().lower()  # Mengubah input ke lowercase
        hash_table = hash_table_by_id

    elif jenis == "nama":
        target = input("Masukkan nama pelanggan: ").strip().lower()  # Mengubah input ke lowercase dan menghapus spasi
        hash_table = hash_table_by_nama

    elif jenis == "tanggal":
        # format harus sama dengan di file, misal: 06-03-2023
        target = input("Masukkan tanggal pembelian (dd-mm-yyyy): ").strip()
        hash_table = hash_table_by_tanggal

    elif jenis == "jumlah":
        target = str(int(input("Masukkan jumlah pembelian (angka): ").strip()))  # Mengonversi jumlah ke string
        hash_table = hash_table_by_jumlah

    elif jenis == "bulanan":
        # format harus sama dengan di file, misal: 06-03-2023
        target = input("Masukkan tanggal pembelian (mm-yyyy): ").strip()
        hash_table = hash_table_by_bulanan

    else:
        print("Jenis pencarian tidak dikenal.")
        return

    # ----- mulai ukur waktu pencarian -----
    start = time.perf_counter()
    
    # Mencari data langsung dari hash table
    results = hash_table.get(target, [])
    elapsed = (time.perf_counter() - start) * 1000  # ms
    # --------------------------------------

    if len(results) == 0:
        print(f"Data tidak ditemukan. Waktu pencarian: {elapsed:.6f} ms")
    else:
        print(f"Ditemukan {len(results)} data. Waktu pencarian: {elapsed:.6f} ms")
        for i, result in enumerate(results, start=1):
            print(f"\nHasil ke-{i}:")
            print(f"  id               : {result['id']}")
            print(f"  nama_pelanggan   : {result['nama_pelanggan']}")
            print(f"  tanggal_pembelian: {result['tanggal_pembelian']}")
            print(f"  jumlah_pembelian : {result['jumlah_pembelian']}")

cari_data()
```
